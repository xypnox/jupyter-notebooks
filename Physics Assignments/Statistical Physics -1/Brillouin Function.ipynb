```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.constants import k,e #in SI
fig, graph = plt.subplots()
```

```python
def coth(x):
    return (np.exp(2*x) + 1)/(np.exp(2*x) - 1)

def B(j, x):
    return (2*j+1)/(2*j) * coth((2*j+1)/2*x)-1/(2*j)*coth(x/2)

```

```python
t = np.arange(-10., 10., 0.1)
```

```python
J = [1/2]

for j in J:
    graph.plot(t, B(j, t), ".r")
graph.grid()
graph.set(
    xlabel = "x",
    ylabel = "Bj",
    title  = "Bj"
)

fig
```
