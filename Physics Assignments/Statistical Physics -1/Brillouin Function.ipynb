```python
import numpy as np
from scipy.constants import k,e #in SI
```

```python
def coth(x):
    return 1/np.tanh(x)

def B(j, x):
    return (2*j+1)/(2*j) * coth((2*j+1)/(2*j)*x)-1/(2*j)*coth(x/(2*j))

def Bjinf(x):
    return coth(x) - 1/x
```

```python
t = np.arange(-10., 10., 0.02)
```

```python
from matplotlib import pyplot as plt
fig, graph = plt.subplots()
J = [
    (1/2, "#3FB8AF", "1/2"),
    (3/2, "#7FC7AF", "3/2"),
    (5/2, "#DAD8A7", "5/2"),
    (45/2, "#FF9E9D", "45/2")
]

for j, fm, label in J:
    graph.plot(t, B(j, t), color=fm, label=label)

    
#Plot for inf
graph.plot(t, Bjinf(t), color="#FF3D7F", label=r"$\infty$")

graph.grid()
graph.legend(title="Value of J")
graph.set(
    xlabel = "x",
    ylabel = r"$B_J(x)$",
    title  = "Brillouin Function"
)
```

```python
# Save the graph (in the end)
fig.savefig('brillouin-function-graph.png', dpi = 300)
```
