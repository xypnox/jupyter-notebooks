```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.constants import k,e #in SI
fig, graph = plt.subplots()

data_points = np.array([
    [  10, 1.15],
    [  20, 0.88],
    [  30, 0.53],
    [  40, 0.32],
    [  50, 0.23],
    [  60, 0.06],
    [  70, 0.12],
    [ 100, 0.39],
    [ 200, 0.90],
    [ 240, 1.00],
    [ 280, 1.13],
    [ 300, 1.16],
    [ 350, 1.19],
    [ 400, 1.24],
    [ 480, 1.38],
    [ 560, 1.35],
    [ 700, 1.40],
    [1000, 1.43],
    [2000, 1.45],
    [5000, 1.45]
])

x, y = data_points.T
```

```python
graph.semilogx(x, y, 'b+')
graph.grid()
graph.set(
    xlabel = "Frequency (Hz)",
    ylabel = "Gain",
    title  = "Band Reject filter"
)

fig.savefig('bandrejectfilter.png', dpi = 300)
fig
```
