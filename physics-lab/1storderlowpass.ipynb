```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.constants import k,e #in SI
fig, graph = plt.subplots()

data_points = np.array([
    [100, 5.56],
    [200, 5.5],
    [500, 5.5],
    [750, 5.33],
    [1000, 5.33],
    [2.5e3, 5],
    [5e3, 4.33],
    [7.5e3, 3.73],
    [10e3, 3.13],
    [20e3, 2.83],
    [30e3, 2.40],
    [40e3, 1.33],
    [50e3, 1.23],
    [75e3, 1.00],
    [100e3, 0.73],
    [200e3, 0.46],
    [500e3, 0.33],
    [1000e3, 0.32],
])

x, y = data_points.T
```

```python
graph.semilogx(x, y, 'b+')
graph.grid()
graph.set(
    xlabel = "Frequency (Hz)",
    ylabel = "Gain",
    title  = "1st order low pass filter"
)

fig.savefig('1storderlowpass.png', dpi = 300)
fig
```
