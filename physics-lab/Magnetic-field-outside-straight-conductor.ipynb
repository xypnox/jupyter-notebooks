```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from scipy.constants import k,e #in SI
(fig, graph) = plt.subplots()
plt.figure(figsize=(60, 60))
plt.style.use('seaborn-muted')

data_points = np.array([
     #I(A)  #B(mT)
    [ 7.00,  0.1 ],
    [14.60,  0.2 ],
    [20.20,  0.3 ],
    [27.10,  0.4 ],
    [34.00,  0.5 ],
    [41.10,  0.6 ],
    [48.00,  0.7 ],
    [55.00,  0.8 ]
])

x, y = data_points.T
```

```python
line = np.polyfit(x, y, 1)
print("slope = ", line[0])
```

```python
# print(plt.style.available)
```

```python
graph.plot(x, y,  ".C2", markersize=15, label="Data")
graph.plot(x, line[0]*x + line[1], "C0", label="Fit")
graph.grid()
graph.set(title='Magnetic Field (mT) vs in Current (A) at a fixed distance',
       ylabel='Magnetic Field (mT)',
       xlabel='Current (A)'
)
graph.text(30, 0.2, "m = %.4f" %line[0])
graph.legend()

fig.savefig('Magnetic-field-outside-straight-conductor_graph1.png', dpi = 300)
fig
```
