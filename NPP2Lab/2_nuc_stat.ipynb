```python
import csv
```

```python
counts = []
```

```python
with open('exp1_2.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for i, row in enumerate(spamreader):
        print(row, i)
        if i > 0:
            counts.append(int(row[1]))

counts
```

```python
import numpy as np
from scipy.constants import k,e #in SI
```

```python
from collections import Counter

counters = Counter(counts)
```

```python
c_av = sum(counts)/len(counts)

c_m_c_av = [x - c_av for x in counts]

print(c_av, c_m_c_av)
```

```python
sig = c_av**(0.5)
vals = [round(x/sig * 2) / 2 for x in c_m_c_av]
val_count = Counter(vals)
val_count, sig, c_av
```

```python
import numpy as np
import matplotlib.pyplot as plt

# the histogram of the data
n, bins, patches = plt.hist(vals, [x/2 + 0.25 for x in range(-8, 8)], rwidth=0.25, facecolor='c', alpha=0.75)


plt.xlabel('N - Nav / sig')
plt.ylabel('Frequency')
plt.title('Distribution for background')
plt.grid(True)
plt.savefig('exp_1_2.png', dpi=300)
```

```python
import csv

# field names 
fields = ['Ni', 'Ni - Nav', 'Ni-Nav/sig', 'Ni-Nav/sig rounded']   
# data rows of csv file 
rows = []

for i in counts:
    row = [i]
    row.append(round(i - c_av, 2))
    row.append(round((i-c_av)/sig, 2))
    row.append(round((i-c_av)/sig * 2)/2)
    rows.append(row)

with open('e_2.csv', 'w+') as f:
    # using csv.writer method from CSV package
    write = csv.writer(f)
    write.writerow(fields)
    write.writerows(rows)
```
